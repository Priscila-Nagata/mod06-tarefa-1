```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sinasc_ro_2019 = pd.read_csv('sinasc_RO_2019.csv')

sinasc_ro_2019.head()
```

```python
sinasc_ro_2019.shape
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x='APGAR1', data=sinasc_ro_2019)

plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.title('Contagem de APGAR1')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x='APGAR5', data=sinasc_ro_2019)

plt.xlabel('APGAR5')
plt.ylabel('Contagem')
plt.title('Contagem de APGAR5')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.countplot(x='SEXO', data=sinasc_ro_2019)

plt.xlabel('SEXO')
plt.ylabel('Contagem')
plt.title('Contagem de SEXO')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.countplot(x='GESTACAO', data=sinasc_ro_2019)

plt.xticks(rotation=60, ha='right')

plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
plt.title('Contagem de GESTACAO')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sinasc_ro_2019.loc[sinasc_ro_2019['CONSULTAS'] == 1, 'Consultas_Desc'] = '1 - Nenhuma'
sinasc_ro_2019.loc[sinasc_ro_2019['CONSULTAS'] == 2, 'Consultas_Desc'] = '2 - De 1 a 3'
sinasc_ro_2019.loc[sinasc_ro_2019['CONSULTAS'] == 3, 'Consultas_Desc'] = '3 - De 4 a 6'
sinasc_ro_2019.loc[sinasc_ro_2019['CONSULTAS'] == 4, 'Consultas_Desc'] = '4 - 7 e mais'
sinasc_ro_2019.loc[sinasc_ro_2019['CONSULTAS'] == 9, 'Consultas_Desc'] = '5 - Ignorado'

sinasc_ro_2019.head(3)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc_ro_2019 )

plt.xlabel('CONSULTAS')
plt.ylabel('MÉDIA_APGAR5')
plt.title('APGAR5 - MÉDIA POR CONSULTA')
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

plt.xticks(rotation=60, ha='right')

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc_ro_2019)
sns.countplot(ax = axes[0, 1], x='CONSULTAS', data=sinasc_ro_2019)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc_ro_2019)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc_ro_2019)

plt.subplots_adjust(wspace=0.25, hspace=0.25)
```
